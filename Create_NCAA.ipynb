## Imports

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
import os,sys
import requests
import pandas as pd
from pandas import DataFrame, read_csv
import scipy as sc
import numpy as np
import statsmodels.formula.api as sm
import re
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

## Scraping the data

```python
arr = []
b = []
scrape = pd.DataFrame()
for x in range(2003,2020):
    for i in range(1,20):
        url = "https://basketball.realgm.com/ncaa/stats/" + str(x) + "/Totals/Qualified/All/Season/All/points/desc/" + str(i)
        html = urlopen(url)
        soup = BeautifulSoup(html)
        soup.findAll('tr', limit=2)
        headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
        headers = headers[:]
        rows = soup.findAll('tr')[1:]
        player_stats = [[td.getText() for td in rows[i].findAll('td')] for i in range(len(rows))]
        stats = pd.DataFrame(player_stats, columns = headers)
        stats = stats.assign(Year = x)
        stats.Player = stats.Player.astype(str)  
        del stats['#']
        b = []
        for y in range(0,len(rows)):
            for a in rows[y].find_all('a'):
                if(a['href'] != None):
                    b.append(a['href'])
                else:
                    b.append(None)
        stats['ID'] = b
        scrape = pd.concat([scrape, stats])
scrape = scrape.reindex(columns=['Year', 'ID'] +list(scrape.columns[:-2]))
scrape
```

## Getting the columns and the names of the players

```python
columnsTable = scrape.columns
coulmns = (columnsTable.to_series())
Names = scrape.drop_duplicates(subset='ID')
Names = Names.iloc[:, 1].tolist()
Names
```

```python
newTableCoulmns = ['ID', 'Player', 'Years', 'From', 'To', 'Team', 'GP', 'MIN', 'FGM', 'FGA', 'FG%',
       '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'TOV', 'PF', 'ORB', 'DRB',
       'REB', 'AST', 'STL', 'BLK', 'PTS']
```

# Creating the total of each player in the NCAA

```python
playersCareerTOT = pd.DataFrame()
for x in range(0, len(Names)):
    currStat = []
    currID = Names[x]
    currPlayer = scrape[scrape['ID'] == currID]
#   Player ID
    currStat.append(currID)
#   Player Name
    currStat.append(currPlayer['Player'].values[0])
#   Total Seasons, Starting and Ending
    currStat.append(currPlayer.describe()['Year'][0])
    currStat.append(currPlayer.describe()['Year'][3])
    currStat.append(currPlayer.describe()['Year'][7])
#   Player College
    currStat.append(currPlayer.describe(include='all')['Team'][2])
    for i in range(4, len(columnsTable)):
        currStat.append(currPlayer[columnsTable[i]].sum())
    currStat = pd.DataFrame(currStat)
    currStat = pd.DataFrame(currStat.transpose())
    currStat.columns = newTableCoulmns
    currStat['FG%'] = (currStat['FGM'] / currStat['FGA']) * 100
    currStat['3P%'] = (currStat['3PM'] / currStat['3PA']) * 100
    currStat['2P%'] = ((currStat['FGM'] - currStat['3PM']) / (currStat['FGA'] - currStat['3PA']) * 100)
    currStat['eFG%'] = ((currStat['FGM'] + 0.5 * currStat['3PM']) / currStat['FGA']) * 100
    currStat['FT%'] = (currStat['FTM'] / currStat['FTA']) * 100
    playersCareerTOT = pd.concat([playersCareerTOT,currStat])
playersCareerTOT
```

```python
playersCareerTOT = playersCareerTOT.reset_index()
del playersCareerTOT['index']
playersCareerTOT
```

# Creating the average dataframe for the players in the NCAA

```python
playersCareerAVG = pd.DataFrame()
for x in range(0, len(Names)):
    currStat = []
    currID = Names[x]
    currPlayer = scrape[scrape['ID'] == currID]
#   Player ID
    currStat.append(currID)
#   Player Name
    currStat.append(currPlayer['Player'].values[0])
#   Total Seasons, Starting and Ending
    currStat.append(currPlayer.describe()['Year'][0])
    currStat.append(currPlayer.describe()['Year'][3])
    currStat.append(currPlayer.describe()['Year'][7])
#   Player College
    currStat.append(currPlayer.describe(include='all')['Team'][2])
    currStat.append(currPlayer['GP'].sum())
    for i in range(5, len(columnsTable)):
        currStat.append(round(currPlayer[columnsTable[i]].sum() / currPlayer['GP'].sum(),3))
    currStat = pd.DataFrame(currStat)
    currStat = pd.DataFrame(currStat.transpose())
    currStat.columns = newTableCoulmns
    currStat['FG%'] = (currStat['FGM'] / currStat['FGA']) * 100
    currStat['3P%'] = (currStat['3PM'] / currStat['3PA']) * 100
    currStat['2P%'] = ((currStat['FGM'] - currStat['3PM']) / (currStat['FGA'] - currStat['3PA']) * 100)
    currStat['eFG%'] = ((currStat['FGM'] + 0.5 * currStat['3PM']) / currStat['FGA']) * 100
    currStat['FT%'] = (currStat['FTM'] / currStat['FTA']) * 100
    playersCareerAVG = pd.concat([playersCareerAVG,currStat])
playersCareerAVG
```
