```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
import os
import requests 

import pandas as pd
from pandas import DataFrame, read_csv
import scipy as sc
import numpy as np
import seaborn as sns

import statsmodels.formula.api as sm
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing

import matplotlib.pyplot as plt 
plt.style.use('ggplot')
%matplotlib inline  
plt.rcParams['figure.figsize'] = (10, 6) 
```

```python
careerAVGPlayer = pd.read_csv('Data/CSV Files/Database/careerAVGPlayer.csv')
```

```python
careerAVGNoCollege = careerAVGPlayer[careerAVGPlayer['Colleges'] == 'None']
careerAVGPlayer = careerAVGPlayer[careerAVGPlayer['PTS'] > 10]
careerAVGPlayer.head()
```

```python
careerAVGNoCollege
```

```python
prizedPlayersNoCollege = careerAVGNoCollege.iloc[:,:24]
prizedPlayersNoCollege['Total'] = prizedPlayersNoCollege['MVP'] + prizedPlayersNoCollege['Finals MVP'] + prizedPlayersNoCollege['All-Star MVP'] + prizedPlayersNoCollege['Most Improved'] + prizedPlayersNoCollege['Sixth Man'] + prizedPlayersNoCollege['Defence Player']+ prizedPlayersNoCollege['Rookie'] + prizedPlayersNoCollege['All Tot'] + prizedPlayersNoCollege['All-Star'] + prizedPlayersNoCollege['D Tot']
prizedPlayersNoCollege = prizedPlayersNoCollege[prizedPlayersNoCollege['Total'] > 0]
prizedPlayersNoCollege
```

```python
collStats = []
collegeSuccess = pd.DataFrame()
collegeSuccessCol = ['College', 'Total Players', 'All-NBA', 'All-Star', 'Defence Player', 'Sixth Man', 'MVP', 'Finals MVP', 'Most Improved', 'Rookie', 'Total Prized Players']
collStats = []
tot = 0
currColl = careerAVGNoCollege[careerAVGNoCollege.Colleges == 'None']
#     College Name
collStats.append('None')
collStats.append(careerAVGNoCollege.shape[0])
#     Precent of All-Star
currAllStar = currColl[currColl['All Tot'] > 0].shape[0]
collStats.append(currAllStar)
for i in range(3, len(collegeSuccessCol) - 1):
#     Precent of All-NBA Team
    currStats = currColl[currColl[collegeSuccessCol[i]] > 0].shape[0]
    collStats.append(currStats)
tot = prizedPlayersNoCollege.shape[0]
collStats.append(tot)
collStats = pd.DataFrame(collStats)
collStats = pd.DataFrame(collStats.transpose())
collStats.columns = collegeSuccessCol
collegeSuccess = pd.concat([collegeSuccess, collStats])
# collegeSuccess['Tot'] = collegeSuccess['All Star'] + collegeSuccess['All-NBA'] + collegeSuccess['Defence']
collegeSuccess = collegeSuccess.sort_values('Total Prized Players', ascending=False)
# collegeSuccess.set_index("College", inplace = True) 
collegeSuccess.head()
```

```python
collStats1 = []
collegeSuccess1 = pd.DataFrame()
collegeSuccessCol1 = ['College', 'Year/PLR', 'All-NBA', 'All-Star', 'Defence Player', 'Sixth Man', 'MVP', 'Finals MVP', 'Most Improved', 'Rookie', 'Total Awards']
collStats1 = []
tot1 = 0
currColl1 = careerAVGNoCollege[careerAVGNoCollege.Colleges == 'None']
#     College Name
collStats1.append('None')
#     Years For Player
numOfPlayers = currColl1.shape[0]
collStats1.append(currColl1['Years'].sum() / numOfPlayers)
#     Precent of All-Star
currAllStar1 = currColl1[currColl1['All Tot'] > 0]['All Tot'].sum()
collStats1.append((int)(currAllStar1))
tot1 = tot1 + currAllStar1
for i in range(3, len(collegeSuccessCol1) - 1):
#     Precent of All-NBA Team
    currStats1 = currColl1[currColl1[collegeSuccessCol1[i]] > 0][collegeSuccessCol1[i]].sum()
    collStats1.append((int)(currStats1))
    tot1 = tot1 + currStats1
collStats1.append(tot1)
collStats1 = pd.DataFrame(collStats1)
collStats1 = pd.DataFrame(collStats1.transpose())
collStats1.columns = collegeSuccessCol1
    
collegeSuccess1 = pd.concat([collegeSuccess1, collStats1])
# collegeSuccess['Tot'] = collegeSuccess['All Star'] + collegeSuccess['All-NBA'] + collegeSuccess['Defence']
addPrizedPlayer = collegeSuccess[['College', 'Total Prized Players']]
collegeSuccess1 = pd.merge(collegeSuccess1, addPrizedPlayer, on='College', how='inner')
rating = []
totalPlayers = pd.Series(collegeSuccess1['Total Prized Players'])
for r in range(0 ,len(collegeSuccess1)):
    if (totalPlayers.values[r] > 0):
        rating.append(round((collegeSuccess1['Sixth Man'].values[r] * 8 + collegeSuccess1['All-NBA'].values[r] * 15 + collegeSuccess1['All-Star'].values[r] * 13 + collegeSuccess1['Defence Player'].values[r] * 10
                            + collegeSuccess1['MVP'].values[r] * 20 + collegeSuccess1['Finals MVP'].values[r] * 24 + collegeSuccess1['Most Improved'].values[r] * 4
                            + collegeSuccess1['Rookie'].values[r] * 8) / totalPlayers.values[r], 2))
    else:
        rating.append(0)
collegeSuccess1['Rating'] = rating
collegeSuccess1 = collegeSuccess1.sort_values('Rating', ascending=False)
# collegeSuccess1.set_index("College", inplace = True) 
collegeSuccess1.head()
```

```python
collegeSuccess1['ScaledRating'] = (collegeSuccess1['Rating']) / 50;
collegeSuccess1['ScaledRating'] =  round(collegeSuccess1['ScaledRating'] , 2)
collegeSuccess1
```

```python
collegeSuccessCopy = collegeSuccess[['College', 'Total Players', 'Total Prized Players']]
collegeSuccess1Copy = collegeSuccess1[['College', 'Year/PLR', 'Total Awards', 'ScaledRating']]
```

```python
noCollegeDF = pd.merge(collegeSuccessCopy, collegeSuccess1Copy, on='College', how='inner')
noCollegeDF
```
