## MVP - Most Valuable Player

```python
url = 'https://www.basketball-reference.com/awards/mvp.html'
html = urlopen(url)

soup = BeautifulSoup(html)
soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')][2:]
rows = soup.findAll('tr')[2:]
links = [[a for a in rows[i].findAll('a')]
        for i in range(len(rows))]
playersHref = '(?:href=")([^"]+)'
    # linksCopy = (str)(links)
ids = re.findall(playersHref, (str)(links))
idsList = []
for i in range(0, len(ids) - 1):
    if('players' in ids[i]):
        currID = ids[i].split('/')[3]
        idsList.append(currID.split('.')[0])
player_stats = [[td.getText() for td in rows[i].findAll('td')][:3]
        for i in range(len(rows))]
stats = pd.DataFrame(player_stats, columns = headers)
playerMVP = idsList[:38]
playerMVP
```

## All-NBA Team

```python
url = 'https://www.basketball-reference.com/awards/all_league.html'

html = urlopen(url)

soup = BeautifulSoup(html)
#     print(soup)
    # use findALL() to get the column headers
soup.findAll('tr', limit=2)

    # use getText()to extract the text we need into a list
# headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
headers = ['Year', 'Team' ,'Center', 'Forward', 'Forward', 'Guard', 'Guard']
rows = soup.findAll('tr')[1:]
links = [[a for a in rows[i].findAll('a')]
        for i in range(len(rows))]
playersHref = '(?:href=")([^"]+)'
    # linksCopy = (str)(links)
ids = re.findall(playersHref, (str)(links))
idsList = []
for i in range(0, len(ids) - 1):
    if('players' in ids[i]):
        currID = ids[i].split('/')[3]
        idsList.append(currID.split('.')[0])
# print(idsList)
year = 2019
numOfPlayer = 0
allTeam = pd.DataFrame()
for numOfYears in range(0,39):
    firstTeam = []
    firstTeam.append(year)
    firstTeam.append("1st")
    for y in range(0, 5):
        firstTeam.append(idsList[numOfPlayer])
        numOfPlayer = numOfPlayer + 1
    firstTeam = pd.DataFrame(firstTeam)
    firstTeam = pd.DataFrame(firstTeam.transpose())
    firstTeam.columns = headers
    allTeam = pd.concat([allTeam,firstTeam])
    secondTeam = []
    secondTeam.append(year)
    secondTeam.append("2nd")
    for y in range(0, 5):
        secondTeam.append(idsList[numOfPlayer])
        numOfPlayer = numOfPlayer + 1
    secondTeam = pd.DataFrame(secondTeam)
    secondTeam = pd.DataFrame(secondTeam.transpose())
    secondTeam.columns = headers
    allTeam = pd.concat([allTeam,secondTeam])
    thirdTeam = []
    thirdTeam.append(year)
    thirdTeam.append("3rd")
    for y in range(0, 5):
        thirdTeam.append(idsList[numOfPlayer])
        numOfPlayer = numOfPlayer + 1
    thirdTeam = pd.DataFrame(thirdTeam)
    thirdTeam = pd.DataFrame(thirdTeam.transpose())
    thirdTeam.columns = headers
    allTeam = pd.concat([allTeam,thirdTeam])
    year = year - 1
```
